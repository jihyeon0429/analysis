###### 키워드 추가

```python
addwords = ['1999','2010','2011','2012','2013','2014','2015','2016','2017','2018','2019','2020',\
            '2021','2022','2023','10·30 대책','9.1대책','9·2 대책','스카이72','리모델링','이면계약',\
            '3+3 개정안','3+3년 개정안','5억 이하','전세가','심각','젋은층','매년','전셋돈','세입자','무조건','홍남기','문의']
```

```python
nouns_spr = nouns_spr + addwords
nouns_spr = set(nouns_spr)
nouns_spr = list(nouns_spr)
```

##### 삽질

```python
dic = {}
for i in nouns_unique:
    a = []    
    for j in nouns_spr:
        if j in i:
            a.append(j)
    a.sort(key=len)
    if a == []:
        a.append(i)
    
    for k in a:
        for l in a:
            if (k in l) and (k != l):
                idx = a.index(k)
                a.pop(idx)
    dic[i] = a
```

##### 명사조각들 카운트 구하기

```python
#추출한 명사조각들이 SPO리스트에서(중복제거X, 1012개 문장) 몇번씩 출현했는지 담기
count = {}
for i in nouns_spr:
    cnt = 0
    for j in nouns_list:
        if i in j:
            cnt+=1
    count[i] = cnt
```

```python
sorted(count.items(), key=lambda x:x[1], reverse=True)
```

#### sum(추출된 명사 카운트)

```python
#단어가 길수록(명사조각이 많이 포함되어 있을수록) 출현했을수록 count 증가
weight = {}
for i in nouns_unique:
    w = 0
    for j in nouns_spr:
        if j in i:
            w += count[j]
    if w == 0:
        weight[i] = 0.1
    else:    
        weight[i] = w      
```

```python
sorted(weight.items(), key=lambda x:x[1], reverse=True)
```

#### sum(추출된 명사 카운트) / 명사조각 개수

```python
# 각 명사조각의 가중치를 반영하여 노드의 가중치 결정
weight = {}
for i in nouns_unique:
    w = 0
    num = 0
    for j in nouns_spr:        
        if j in i:
            w += count[j]
            num+=1
            #print(i,'/',j)
    try:        
        w = round(w/num,2)
    except ZeroDivisionError:
        w = w
    if w == 0:
        weight[i] = 0.1
    else:    
        weight[i] = w        
```

```python
sorted(weight.items(), key=lambda x:x[1], reverse=True)
```

```python
for i in nouns_unique:
    print(i,weight[i],'*')
    for j in nouns_spr:
        if j in i:
            print(j, count[j])
    print('---------------')
```

####  중복 카운트 이슈 제거

- 보다 구체적인 키워드만 선택 ('전세', '전세난' 중 '전세난'만 선택)

```python
dic = {}
for i in nouns_unique:
    a = []    
    for j in nouns_spr:
        if j in i:
            a.append(j)
    a.sort(key=len)
    if a == []:
        a.append(i)
    
    for k in a:
        for l in a:
            if (k in l) and (k != l):
                idx = a.index(k)
                a.pop(idx)
    dic[i] = a
```

```python
weight = {}
for i in nouns_unique:
    try:
        w = 0
        cnt = 0
        for j in dic[i]:
            w += count[j]
            cnt += 1
        try: 
            w = round(w/cnt, 2)
        except ZeroDivisionError:
            w = w
    except KeyError:
        w = 0.1
    weight[i] = w
```

```python
sorted(weight.items(), key=lambda x:x[1], reverse=True)
```
